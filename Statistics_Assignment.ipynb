1.What is statistics, and why is it important?


*   Statistics is the study of data collection, analysis, interpretation, and presentation. It is important because it helps in decision-making, understanding trends, and making predictions in various fields such as business, healthcare, and science.



2.What are the two main types of statistics?


*   Descriptive Statistics: Summarizes and organizes data (e.g., mean, median, mode, standard deviation).
*Inferential Statistics: Draws conclusions and makes predictions about a population based on a sample.



3.What are descriptive statistics?


*   Descriptive statistics provide numerical summaries of data, such as measures of central tendency (mean, median, mode) and measures of variability (variance, standard deviation, range).



4.What is inferential statistics?    
  

* Inferential statistics use sample data to make conclusions about a larger population using techniques such as hypothesis testing, regression analysis, and confidence intervals.





5.What is sampling in statistics?


*   Sampling is selecting a subset of data from a population to analyze trends and make predictions without studying the entire population.


6.What are the different types of sampling methods?


*   Random Sampling: Each individual has an equal chance of selection.
*tratified Sampling: Population is divided into groups, and samples are taken proportionally.
*Systematic Sampling: Every nth item is selected from a list.
*Cluster Sampling: The population is divided into clusters, and entire clusters are sampled.
*Convenience Sampling: Selection is based on ease of access.



7.What is the difference between random and non-random sampling?


*   Random Sampling ensures every individual has an equal chance of selection, reducing bias.
*Non-Random Sampling involves selection based on criteria such as convenience or judgment, leading to possible bias.



8.Define and give examples of qualitative and quantitative data.


*   Qualitative Data: Non-numeric data (e.g., colors, names, categories).
*Quantitative Data: Numeric data (e.g., height, weight, temperature).



9.What are the different types of data in statistics?


*   Nominal: Categories without order (e.g., colors, gender).
*Ordinal: Categories with a meaningful order but unknown differences (e.g., rankings, satisfaction levels).
*Interval: Numeric data with known differences but no true zero (e.g., temperature in Celsius).
+Ratio: Numeric data with a true zero (e.g., weight, height, income).



10.What is the measure of central tendency?



*  It describes the center of a dataset, including mean, median, and mode.




11.Define mean, median, and mode.


*   Mean: The average of a dataset.
*Median: The middle value when data is sorted.
*Mode: The most frequently occurring value.



12.What is variance, and how is it calculated?


*   Variance measures the spread of data from the mean. It is calculated as the average of squared differences from the mean.



13.What is standard deviation, and why is it important?


*   Standard deviation is the square root of variance and indicates how spread out the data points are.



14.What is skewness in a dataset?


*   Skewness measures the asymmetry of the data distribution.



15.Define and explain kurtosis.


* Kurtosis measures the "tailedness" of a distribution, indicating whether data has heavy or light tails compared to a normal distribution.


16.What does correlation measure in statistics?


*   Correlation measures the relationship between two variables, indicating how one variable changes in relation to another.

```python
#1.How do you calculate the mean, median, and mode of a dataset
import numpy as np
from scipy import stats

# Sample dataset
data = [10, 20, 20, 30, 40, 50, 60, 70, 80, 90]

# Mean
mean_value = np.mean(data)

# Median
median_value = np.median(data)

# Mode
mode_value = stats.mode(data)

print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Mode: {mode_value.mode}")
```

```python
2#Write a Python program to compute the variance and standard deviation of a dataset
import numpy as np

# Sample dataset
data = [10, 20, 30, 40, 50, 60, 70, 80, 90]

# Variance
variance = np.var(data, ddof=0)  # Population variance (ddof=0)
sample_variance = np.var(data, ddof=1)  # Sample variance (ddof=1)

# Standard Deviation
std_dev = np.std(data, ddof=0)  # Population standard deviation
sample_std_dev = np.std(data, ddof=1)  # Sample standard deviation

print(f"Population Variance: {variance}")
print(f"Sample Variance: {sample_variance}")
print(f"Population Standard Deviation: {std_dev}")
print(f"Sample Standard Deviation: {sample_std_dev}")
```

```python
7.# Calculate skewness and kurtosis of a dataset using Python libraries.
import numpy as np
from scipy.stats import skew, kurtosis
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

# Generate a right-skewed dataset
data = np.random.exponential(scale=2, size=1000)

# Calculate skewness and kurtosis
skewness = skew(data)
kurt = kurtosis(data)

print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurt}")

# Plot the histogram to visualize skewness
plt.figure(figsize=(8,5))
sns.histplot(data, bins=30, kde=True, color="blue")
plt.title("Histogram of Skewed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
8#Generate a dataset and demonstrate positive and negative skewness
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generate a positively skewed dataset (Right-Skewed)
positive_skew_data = np.random.exponential(scale=2, size=1000)

# Generate a negatively skewed dataset (Left-Skewed)
negative_skew_data = -np.random.exponential(scale=2, size=1000)

# Calculate skewness
positive_skewness = stats.skew(positive_skew_data)
negative_skewness = stats.skew(negative_skew_data)

print(f"Skewness of Positively Skewed Data: {positive_skewness:.4f}")
print(f"Skewness of Negatively Skewed Data: {negative_skewness:.4f}")

# Plot histograms
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Positive skew
axes[0].hist(positive_skew_data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
axes[0].set_title("Positively Skewed Distribution (Right-Skewed)")
axes[0].set_xlabel("Values")
axes[0].set_ylabel("Frequency")

# Negative skew
axes[1].hist(negative_skew_data, bins=30, color='salmon', edgecolor='black', alpha=0.7)
axes[1].set_title("Negatively Skewed Distribution (Left-Skewed)")
axes[1].set_xlabel("Values")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
12#Implement and compare simple random sampling and systematic sampling
import numpy as np
import pandas as pd

# Generate a dataset of 100 elements
np.random.seed(42)
data = np.random.randint(1, 100, 100)
df = pd.DataFrame(data, columns=["Values"])

# Simple Random Sampling (select 10 random samples)
random_sample = df.sample(n=10, random_state=42)

# Systematic Sampling (select every k-th element, k = len(data)//10)
k = len(df) // 10
systematic_sample = df.iloc[::k]

# Display samples
print("Simple Random Sampling:\n", random_sample)
print("\nSystematic Sampling:\n", systematic_sample)
```

```python
13#Calculate the mean, median, and mode of grouped data
import numpy as np
import pandas as pd
from scipy import stats

# Sample grouped data (values represent frequency distribution)
data = {
    "Class Interval": ["1-10", "11-20", "21-30", "31-40", "41-50"],
    "Frequency": [5, 8, 15, 10, 7]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Calculate midpoint for each class interval
df["Midpoint"] = df["Class Interval"].apply(lambda x: (int(x.split('-')[0]) + int(x.split('-')[1])) / 2)

# Compute mean
mean = np.average(df["Midpoint"], weights=df["Frequency"])

# Compute median (approximation using cumulative frequency)
df["Cumulative Frequency"] = df["Frequency"].cumsum()
n = df["Frequency"].sum()
median_class = df[df["Cumulative Frequency"] >= n/2].iloc[0]
L = int(median_class["Class Interval"].split('-')[0])  # Lower boundary of median class
f = median_class["Frequency"]  # Frequency of median class
F = df[df["Cumulative Frequency"] < n/2]["Cumulative Frequency"].max() if not df[df["Cumulative Frequency"] < n/2].empty else 0
h = int(median_class["Class Interval"].split('-')[1]) - L  # Class width
median = L + ((n/2 - F) / f) * h

# Compute mode (approximation using modal class formula)
modal_class = df.iloc[df["Frequency"].idxmax()]
L_mode = int(modal_class["Class Interval"].split('-')[0])
f1 = modal_class["Frequency"]
f0 = df.iloc[df["Frequency"].idxmax() - 1]["Frequency"] if df["Frequency"].idxmax() > 0 else 0
f2 = df.iloc[df["Frequency"].idxmax() + 1]["Frequency"] if df["Frequency"].idxmax() < len(df) - 1 else 0
mode = L_mode + ((f1 - f0) / ((2 * f1) - f0 - f2)) * h if f1 > f0 and f1 > f2 else None

# Print results
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Mode: {mode:.2f}" if mode else "Mode: No distinct mode")
```

```python
14# Simulate data using Python and calculate its central tendency and dispersion
import numpy as np
import scipy.stats as stats

# Simulate a dataset (normal distribution)
np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=1000)  # Mean=50, Std Dev=15, 1000 samples

# Calculate central tendency measures
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data, keepdims=True)[0][0]

# Calculate dispersion measures
variance = np.var(data, ddof=1)  # Sample variance
std_dev = np.std(data, ddof=1)   # Sample standard deviation
range_value = np.ptp(data)       # Range (max - min)

# Print results
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Mode: {mode:.2f}")
print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Range: {range_value:.2f}")
```
